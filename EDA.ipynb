```python
# -*- coding: utf-8 -*-
"""
Data Analysis Script - VS Code Friendly
@author: mohammed zuber
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pylab

# -----------------------------
# Load Education Dataset
# -----------------------------
education = pd.read_csv(r"C:\Users\mohammed zuber\Downloads\InClass_DataPreprocessing_datasets\InClass_DataPreprocessing_datasets\education.csv")
print("Education Data Preview:\n", education.head())

# -----------------------------
# Basic Statistics
# -----------------------------
print("\n--- Work Experience ---")
print("Mean:", education.workex.mean())
print("Median:", education.workex.median())
print("Mode (Pandas):", education.workex.mode()[0])
print("Mode (Scipy):", stats.mode(education.workex, keepdims=True).mode[0])
print("Variance:", education.workex.var())
print("Std Dev:", education.workex.std())
print("Range:", education.workex.max() - education.workex.min())
print("Skewness:", education.workex.skew())
print("Kurtosis:", education.workex.kurt())

print("\n--- GMAT ---")
print("Mean:", education.gmat.mean())
print("Median:", education.gmat.median())
print("Mode (Pandas):", education.gmat.mode()[0])
print("Mode (Scipy):", stats.mode(education.gmat, keepdims=True).mode[0])
print("Variance:", education.gmat.var())
print("Std Dev:", education.gmat.std())
print("Range:", education.gmat.max() - education.gmat.min())
print("Skewness:", education.gmat.skew())
print("Kurtosis:", education.gmat.kurt())

# -----------------------------
# Visualization
# -----------------------------

# Bar Plot of GMAT
plt.figure(figsize=(10,5))
plt.bar(height=education.gmat, x=np.arange(len(education.gmat)))
plt.xlabel("Index")
plt.ylabel("GMAT")
plt.title("Bar Plot of GMAT Scores")
plt.show()

# Histogram
plt.figure(figsize=(10,5))
plt.hist(education.gmat, bins=[600,640,680,720,760,780], color='blue', edgecolor='red')
plt.xlabel("GMAT")
plt.ylabel("Frequency")
plt.title("Histogram of GMAT")
plt.show()

plt.figure(figsize=(6,5))
plt.hist(education.workex, bins=6, color='red', edgecolor='black', orientation='horizontal')
plt.xlabel("Frequency")
plt.ylabel("Work Experience")
plt.title("Histogram of Work Experience")
plt.show()

# Boxplots
plt.figure(figsize=(6,5))
plt.boxplot(education.gmat)
plt.title("Boxplot of GMAT")
plt.show()

plt.figure(figsize=(6,5))
plt.boxplot(education.workex)
plt.title("Boxplot of Work Experience")
plt.show()

# Density plots
plt.figure(figsize=(6,5))
sns.kdeplot(education.gmat, bw_adjust=0.5, fill=True, color='red')
plt.title("Density Plot of GMAT")
plt.show()

plt.figure(figsize=(6,5))
sns.kdeplot(education.workex, fill=True, color='green')
plt.title("Density Plot of Work Experience")
plt.show()

# Q-Q Plot
plt.figure(figsize=(6,5))
stats.probplot(education.gmat, dist='norm', plot=pylab)
plt.title("Q-Q Plot of GMAT")
plt.show()

# -----------------------------
# Scatter Plot: Waist vs AT
# -----------------------------
wc = pd.read_csv(r"C:\Users\mohammed zuber\Downloads\wc-at.csv")
print("\nWaist & AT Data Preview:\n", wc.head())

plt.figure(figsize=(6,5))
plt.scatter(wc['Waist'], wc['AT'], color='pink')
plt.xlabel("Waist")
plt.ylabel("AT")
plt.title("Scatter Plot of Waist vs AT")
plt.show()

# Correlation & Covariance
correlation = np.corrcoef(wc.Waist, wc.AT)[0,1]
covariance = np.cov(wc.Waist, wc.AT)[0,1]
print("Correlation (Waist vs AT):", correlation)
print("Covariance (Waist vs AT):", covariance)
```
